# Using Opencv for Cartooning a Picture

We import all the necessary python Libraries for needed this project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

Importing the Image for processing

```python
test_image = cv2.imread("C:\\Users\\Toyin Obasoro\\Pictures\\Aero-EM.png")
```

The test_Image is imported as a BGR model we comvert to RGB

```python
test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(test_image)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(test_image, cmap='gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(test_image)
```

```python
edges_image = cv2.Canny(test_image, 50, 100)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(test_image)
plt.imshow(edges_image)
```

```python
blurred_image = cv2.medianBlur(test_image, 5)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(test_image)
plt.imshow(blurred_image)
```

```python
edges_image = cv2.Canny(blurred_image, 50, 100)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(edges_image)
```
